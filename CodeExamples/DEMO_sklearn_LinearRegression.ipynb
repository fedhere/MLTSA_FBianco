# CLARIFICATION OF SKLEARN LINEAR REGRESSION MODELS BEHAVIOR ON FITTING INTERCEPT

```python
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1)
x = np.arange(1000)
X = x.reshape(-1,1)
y_true = x * 0.3 + 100
noise = np.random.normal(0, 60, 1000)
y = y_true + noise

# with column of ones
X_with_ones = np.hstack((np.ones((X.shape[0], 1)), X))





settings = [("fit_intercept=True X 1D"),
 ("fit_intercept=False X 1D"),
 ("fit_intercept=True X 2D"),
 ("fit_intercept=False X 2D")]

for i,(b,data) in enumerate([(True, X), (False, X), (True, X_with_ones), (False, X_with_ones)]):
  lr = LinearRegression(fit_intercept=b)
  lr.fit(data, y)
  plt.errorbar(x, y, yerr=np.abs(noise), fmt='.', alpha=0.1)
  plt.xlabel("x")
  plt.ylabel("y")
  if data.shape[1] == 1:
    plt.plot(data, lr.predict(data), '-', label=settings[i])
  else:
    plt.plot(data[:, 1], lr.predict(data), '--', label=settings[i])

  print(settings[i], lr.intercept_, lr.coef_)
  plt.legend();
  plt.show()
```
