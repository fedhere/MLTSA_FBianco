```python
import pandas as pd
import matplotlib.pylab as plt
```

```python
videos = pd.read_csv("https://raw.githubusercontent.com/fedhere/FDSFE_FBianco/main/data/USyoutubes.csv")
```

```python
videos_restrict = videos[['views', 'likes', 'dislikes', 'comment_count']]
```

```python
videos_restrict
```

```python
videos_restrict.corr()
```

```python
plt.imshow(videos_restrict.corr(), clim=(-1,1), cmap="RdBu")
plt.xticks([0,1,2,3], ['views', 'likes', 'dislikes', 'comment_count'], rotation=45);
plt.yticks(range(len(videos_restrict.columns)), videos_restrict.columns, rotation=45);
plt.colorbar();
```
